# Aplicación: Directorio de contactos

```python
# -*- coding: utf-8 -*-


#Segundo paso: donde se guardara el primer paso. Generar una clase para los contactos. Sera un contendor de las variables.
class Contact:

    #defino el constructor
    def __init__(self, name, phone, email):
        #inicializo  las variables de instancia
        self.name = name 
        self.phone = phone
        self.email = email
        #con _ las hago privadas, pero en este caso no deberia ser privado porque sino despues no puedo acceder a ellas. deben ser publicas






#Tercer paso: genero una clase para una agenda, donde se guardaran los contactos
class ContactBook:  

    #recordar que todos los metodos de instancia empiezan con self:

    def __init__(self): #dentro del constructor no recibe ningun parametro 
        self._contacts = [] #la inicializo con una lista vacia con los contactos




    #para agregar los contactos genero el metodo add de cada instancia, que recibe las diferentes variables como parametros al nombre, el telefono y el correo:
    def add(self, name, phone, email):

        #genero un contacto: genero una nueva instancia de la clase contacto. Para inicializarla hay que darle los valores en el constructor
        contact = Contact(name, phone, email)

        self._contacts.append(contact) #agrego al final de la lista el contacto)




    #Septimo paso: en el contact book implemento el metodo para mostrar la lista
    def show_all(self):  #no necesita ningun otro parametro porque ya esta mostrando todo 
        for contact in self._contacts:  #con el metodo show_all: itero por cada contacto a lo largo de la lista e imprimo el nombre del contacto
            self._print_contact(contact) #imprimo el contacto en la consola con un metodo privado, llamando al metodo print_contact, que recibe como parametro el contacto 




    def _print_contact(self, contact):
        #denoto que es un metodo privado empezandolo con guion bajo
        print('Nombre: {}'.format(contact.name))
        print('Telefono: {}'.format(contact.phone))
        print('Email: {}'.format(contact.email))
        print('------ * ------ * ------ * ------ * ------ * ------')




    #Noveno paso: implemento el meotod delete
    def delete(self, name):  #itero a lo largo de la lista de contacto y le paso como parametro al nombre, porque si se encuentra al nombre borrare ese elemento de la lista de contactos.
    #para poder borrar elemento se necesita el indice, entonces no se utilizaran directamente los contactos, sino un metodo llamado enumerate(un metodo parecido a iterar, pero que ademas provee el indice y no nada mas el elemento)
        for idx, contact in enumerate(self._contacts):  #por el indice y el contacto en enumerate(con parametro a la lista de contactos),
            if contact.name.lower() == name.lower():    #si el nombre del contacto es igual al nombre que recibo como parametro,
               del self._contacts[idx]                  #entonces lo que se hara es borrar ese elemento de la lista de contactos.    
            break                                 #y si se encuentra el contacto, no hace falta seguir dando vueltas en el loop, entonces salgo. 




    #implementar el metodo search para el Decimo paso:
    def search(self, name):
        for contact in self._contacts:  #iterar dentro de los contactos 
            if contact.name.lower() == name.lower(): #si se da esta igualdad
                self._print_contact(contact)            #se imprime el nombre del contacto  
                break                                   # y ahi termina el loop 
        else:                                  
            self._not_found()            #si se itera todo, para (con break), pero no se encuentró nada, genero el metodo not_found, donde se le aclarara al usuario que no se encontro el contacto pedido

    def _not_found(self):
        print('*******')
        print('¡Contacto no encontrado!')
        print('*******')



        
        
    #Del Onceavo paso: funcion para actualizar contactos 
    def update(self, name, phone, email):
        for contact in self._contacts:
            if contact.name.lower() == name.lower():
                contact.phone = phone
                contact.email = email
                break
        else:
            self._not_found()









# PRIMER PASO: definir las funciones de la agenda como tal, con las opciones 
def run():

    contact_book =  ContactBook() #Cuarto paso: con los valores que da el usuario puedo generar una nueva instancia de la agenda, guardandola dentro de la variable contact_book. Se inicializa llamando al constructor de la clase, que llama a una variable privada llamada contacts que inicializa como una lista donde se guardaran contactos


    while True:
        command = str(input('''
            ¿Que deseas hacer?

            [a]  Añadir contacto
            [ac] Actualizar contacto
            [b]  Buscar contacto
            [e]  Eliminar contacto
            [l]  Listar contactos
            [s]  Salir
        '''))

        
        
        
        if command == 'a':
            print(' ')
            print(' -->  Añadir contacto:')
            #se le pide al usuario datos de contacto
            name = str(input('Nombre del contacto: ')) 
            phone = str(input('Telefono del contacto: '))
            email = str(input('Email del contacto: '))

            #Quinto paso: ya teniendo la agenda llamare a metodos sobre ese objeto:
            contact_book.add(name, phone, email) #metodo añadir, con parametros para el nombre, telefono y correo


            

        elif command == 'ac':
            print(' ')
            print('-->  Actualizar contacto:')

            #Onceavo paso: objetivo: encontrar un contacto, modificar sus propiedades y volverlo a guardar
            name = str(input('Nombre del contacto a actualizar: '))
            print(' ')
            phone = str(input('Telefono nuevo del contacto: '))
            email = str(input('Email nuevo del contacto: '))

            contact_book.update(name, phone, email)




        elif command == 'b':
            print(' ')
            print('-->  Buscar contacto:')

            #Decimo paso: 
            name = str(input('Nombre del contacto a buscar: '))
            print('>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>')

            contact_book.search(name)




        elif command == 'e':
            print(' ')
            print('-->  Eliminar contacto:')

            #Octavo paso: eliminar contacto por nombre
            name = str(input('Nombre del contacto a eliminar: '))

            contact_book.delete(name)  #implemento el metodo delete y le paso el nombre como atributo


            
            
        elif command == 'l':
            print(' ')
            print('# Lista de contactos:')
            print('>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>')

            #Sexto paso: genero un comando para listar los contactos y mostrarlos
            contact_book.show_all()     #uso la instancia de agenda contact_book con el metodo show_all

            

            
        elif command == 's':
            break

            


        else:
            print(' ')
            print('Comando no encontrado.')


if __name__ == '__main__':
    run()
```
