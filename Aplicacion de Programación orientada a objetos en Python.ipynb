    Los objetos nos permiten modelar objetos de la vida real, estos objetos van a tener un estado y métodos(acciones que se pueden generar con este objeto).
De esta forma vamos a poder definir cada parte de nuestro código modelandolas cómo objetos.
Encapsulación es una de las herramientas más poderosas que tienen los programadores, esta nos permite esconder la complejidad del software.

```python
# -*- coding: utf-8 -*-

#Declaro una clase con el keyword class (empezarlas con mayusculas)
class Lamp:
    _LAMPS = ['''               
    #declaro una variable de clase que no depende del estado individual de cada lampara
          .
     .    |    ,
      \   '   /
       ` ,-. '
    --- (   ) ---
         \ /
        _|=|_
       |_____|  Lampara prendida
    ''',
    '''
         ,-.
        (   )
         \ /
        _|=|_
       |_____|  Lampara apagada
    ''']

    #para empezar una clase es necesario definir el metodo __init__ que recibe el parametro self 
    #todos los metoods dentro de las clases en python deben recibir el parametro self que es la propia instancia, para manipular la clase.
    #se ejecuta cuando se genera una nueva instancia de esta clase
    def __init__(self, is_turned_on):
        #defino las variables de la instancia
        self._is_turned_on = is_turned_on #en este caso se define una variable privada que se inicializa como falso

    #declaro un metodo publico
    def turn_on(self):
        self._is_turned_on = True
        self._display_image()
        
        
    #declaro un metodo publico
    def turn_off(self):
        self._is_turned_on = False
        self._display_image()

        
    #declaro un metodo privado que sirve para mostrar la lampara en la consola. Es un metodo privado porque no se quiere exponer. No se desea que los usuarios de la clase puedan modificar la lampara
    def _display_image(self):
        if self._is_turned_on:
            print(self._LAMPS[0])  #si  la lampara esta prendida se muestra el primer elemento de la lista "_Lamps"
        else:
            print(self._LAMPS[1])
            
            
            
       

#defino una funcion que declara una instancia nueva de la clase
def run():
    lamp = Lamp(is_turned_on=False)

    while True:
        command = str(input('''
          ¿Que deseas hacer?

            [p]render
            [a]pagar
            [s]alir
        '''))

        if command == 'p':
            lamp.turn_on()   #si esta prendido se aplica el metodo turn on
        elif command == 'a':
            lamp.turn_off()    #si esta no prendido se aplica el metodo turn of
        else:
            break                

            
       

if __name__ == '__main__':
    run()
```
